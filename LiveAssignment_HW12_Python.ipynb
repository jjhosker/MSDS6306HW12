```python
################################################
## Question 3:  Introduction to pandas:
################################################
print('Question 3:  Introduction to pandas:')
print("")

###### Question 3a:  Import the pandas library â€“ import pandas ######
import pandas
print("     3.a. Imported the pandas library:  import pandas")
print("")

###### Question 3b:  Read the IRIS dataset into iris using pandas.read_csv(). 
######               Data file â€“  ######
print("     3.b. Sucessfully read in Iris.csv file as iris dataset")
#####################################################################
## Change to directory where file is located using your directory name
os.chdir('C:\\Users\\JHosker\\Documents\\PythonProj\\MSDS6306HW12')
## print(os.getcwd())  ## print directory
#####################################################################

## Read in Iris.csv file after changing directory above
iris = pandas.read_csv("Iris.csv")
print("")

###### Question 3c:  Using iris.head(), display the head of the dataset
print("     3.c. Display iris.head() of iris dataset:")
print(iris.head())
print("")

###### Question 3d:  Use DataFrame.drop() to drop the id column
print("     3.d. Use DataFrame.drop() to drop the Id column")
print("          and print iris.head() to show it is removed:")
iris = iris.drop('Id', 1)
print(iris.head())
print("")

###### Question 3e:  Subset dataframe to create a new data frame that
######               includes only the measurements for the setosa species
iris_setosa = iris.loc[iris['Species']=='Iris-setosa']
print("     3.e. We create a subset dataframe (iris_setosa) that includes")
print("          only setosa species and the iris_setosa.head() printed below:")
print(iris_setosa.head())
print("")

###### Question 3f:  Use DataFrame.describe() to get the summary statistics
print("     3.f. We printout iris_setosa.describe() for summary stats below:")
print(iris_setosa.describe())
print("")

###### Question 3g:  Use DataFrame.groupby() to create grouped data frames by 
######               Species and compute summary statistics using 
######               DataFrame.describe()
iris_SpeciesByGrp = iris.groupby(['Species'])
print("     3.g. Use iris.groupby() to create grouped data frames by") 
print("          Species and compute/print summary statistics")
print("          using iris.describe():")
print(iris_SpeciesByGrp.describe())
print("")

###### Question 3h:  Use DataFrame.boxplot() to plot boxplots by Species
print("     3.h. Use iris.boxplot() to plot boxplots by Species categories:") 
## create new column to reduce Species label size
shortlabel = []
# For each row in the column,
for i in iris['Species']:
    if i == 'Iris-setosa':
        shortlabel.append('setosa')
    elif i == 'Iris-versicolor':
        shortlabel.append('versicolor')
    elif i == 'Iris-virginica':
        shortlabel.append('virginica')

iris['SpecLabel'] = shortlabel
iris.boxplot(by='SpecLabel')
print("")
plt.show()
print("")

###### Question 3i:  Plot a scatter matrix plot using the seaborn library. 
######               Use the following to load and plot 
######               i. Import seaborn
######               ii. Seaborn.pairplot(dataframe,by=â€™column_nameâ€™)
import seaborn as sns
print("     3.i. #i. Imported the seaborn library as sns:  import seaborn as sns")
print("")

print("     3.i. #ii. Run sns.pairplot(iris,hue='SpecLabel')")
sns.pairplot(iris,hue='Species')
print("")
plt.show()
print("")

## Here is the code for part (i)
## import seaborn as sns
## sns.pairplot(iris,hue='Species')

## Deliverable is a link to Jupiter notebook file on GitHub.
print("")
urlink_gitHW12= "https://github.com/jjhosker/MSDS6306HW12"
print("Link to Jupiter notebook file on Github repository URL below: ")
print(urlink_gitHW12)
print("View README.md file in Github repository.")
```
